```python
import pulp as p
```

```python
Lp_prob=p.LpProblem("ProductionPlanning",p.LpMaximize)
```

## Desicion variable

```python
x1=p.LpVariable("x1",lowBound=0) #x1>0
x2=p.LpVariable("x2",lowBound=0) #x2>0
```

## Objective Function

```python
Lp_prob+=500*x1+700*x2
```

```python
print(Lp_prob)
```

## Constraints

```python
Lp_prob+=2*x1+4*x2<=80
Lp_prob+=3*x1+2*x2<=60
Lp_prob+=1*x1+3*x2<=45
```

```python
Lp_prob
```

## Solving LPP

```python
status=Lp_prob.solve()
print(p.LpStatus[status])
```

```python
p.value(x1)
```

```python
p.value(x2)
```

```python
p.value(Lp_prob.objective)
```
